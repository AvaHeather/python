```python
# Task Instructions
# 1. Generate a simulated sales dataset, with the following columns: Year, Q1, Q2, Q3, Q4
# 2. Reshape the data, from wide to long, to contain the following columns: Year, Quarters and Sales.
# 3. Calculate the average yearly sales
# 4. Draw a neatly labelled line plot, that shows the average sales for each year.
# 5. Draw a neatly labelled line plot, that shows the quarterly sales for each year.
```

```python
## Import the required libraries
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Set the working environment
os.chdir("/Users/busara/Documents/Github/python")
```

```python
## 1. Generate a simulated sales dataset, with the following columns: County, Year, Q1, Q2, Q3, Q4
Year = list(range(2005,2018,1))
Quarter1 = np.random.choice(range(0,100,2),size =len(Year))
Quarter2 = np.random.choice(range(30,200,3),size =len(Year))
Quarter3 = np.random.choice(range(4,530,9),size =len(Year))
Quarter4 = np.random.choice(range(2,340,16),size =len(Year))

SalesData = pd.DataFrame({'Year':Year,'Quarter1':Quarter1, 'Quarter2': Quarter2, 'Quarter3':Quarter3,'Quarter4':Quarter4})
SalesData

## Save this dataset as a csv file, to be read in R and STATA
SalesData.to_csv("/Users/busara/Documents/Github/SalesData.csv")
```

```python
## 2. Reshape the data, from wide to long, to contain the following columns: County, Year, Quarters and Sales.
SalesData_long = SalesData.melt(id_vars='Year',value_name='Sales',var_name="Quarters",
                                value_vars=['Quarter1','Quarter2','Quarter3','Quarter4'])
SalesData_long
```

```python
## 3. Calculate the average yearly sales.
sales_mean = pd.DataFrame(SalesData_long.groupby(['Year'])['Sales'].mean().reset_index())
sales_mean
```

```python
# 4. Draw a neatly labelled line plot, that shows the average sales for each year.
## Set the size of the figure
plt.figure(figsize=(8,8))
## Generate the factor plot
ax = sns.factorplot(x="Year", y="Sales", data=sales_mean,size=5, aspect=2,color = "red")

##Label the plot appropriately
ax.set(title = "Average Sales Values per Year",xlabel = "Year",ylabel = "Average Sales")
```

```python
# 5. Draw a neatly labelled line plot, that shows the quarterly sales for each year.
## Set the size of the figure
plt.figure(figsize=(10,10))

## Generate the factor plot
ax = sns.factorplot(x="Year", y="Sales", hue ="Quarters", data=SalesData_long,size=5,aspect=2)

##Label the plot appropriately
ax.set(title = "Quarterly Sales Values per Year",
       xlabel = "Year",ylabel = "Sales")
#ax.set_xticklabels(ax.get_xticklabels(), rotation = 30)
```
